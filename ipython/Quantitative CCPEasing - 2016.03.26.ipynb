# Introduction

The recent implementation of the skill point trading system into the EVE economy has far-reaching macroeconomic implications that represent a fundamental shift in CCP's management of the ISK economy. This change was telegraphed by CCP Quant's commentary in the EVE Vegas 2015 Economy Update Report; the inaugural Monthly Economic Report (Feb. 2016) displays the ongoing effects of this decision and also gives some insight into CCP's future monetary priorities. Due to the game's uniquely liberal market structure and the considerable value of the ISK economy, these questions carry with them not only a significant financial impact but an academically useful one as well.

# Background

It is assumed that the reader is at least loosely familiar with Dr. Eyjólfur Guðmundsson's work as CCP's private economist (if not, examples of his work are archived by CCP [LINK](http://cdn1.eveonline.com/community/QEN/QEN_Q4-2010.pdf) and media coverage thereof is widely available). Although the EVE economy has changed significantly since these publications were written, the most salient point that endures may be the growing concern in Eyjólfur's final reports regarding the trading volume of financial products (namely PLEX) as a proportion of the ISK economy; Eyjólfur's fear of a runaway inflationary cycle leads him to suggest in the 2010Q4 report that the contemporary policy of the "EVE Central Bank" should be to decrease the money supply of ISK.

In the years that follow Eyjólfur's departure from CCP in 2011, CCP staff come to the conclusion [LINK](https://community.eveonline.com/news/dev-blogs/price-indices-february-2012/) that despite the presence of tremendous monetary inflation (growth of the ISK money supply), very little actual price inflation has occured (echoing some empirical reviews [LINK](http://krugman.blogs.nytimes.com/2011/10/07/way-off-base-2) of Federal Reserve QE policies). Centering the discussion around a number of price indices, the CCP stance now incorporates the money velocity of ISK to determine that despite the slide in CPI numbers, the economy has not produced a material amount of inflation *or* deflation. The 2012 report ends on an optimistic note, suggesting an uptick in demand-driven growth. Notably,
>The correlation between the velocity of money and the CPI seems to change by the end of 2008.  From January 2007 to October 2008, the correlation between the two series is 0.86, which is quite strong for a non-academic case like this one.  From November 2008 to February 2012 the correlation is -0.01, which is just about as uncorrelated as possible.  So, what happened in November 2008?  PLEX was introduced.  **PLEX seems to increase the velocity of money in the New Eden economy as it "liberates" ISK from people with more money than they know what to do with and gets it to ISK needy people that create and sell PLEX.** [emphasis added]

>...

> An important factor in this was probably the absence of a trusted banking sector, which could offer interest and distribute the unused ISK to those in need of funds. 

In other words, CCP realizes that the seeming glut of ISK has been hoarded in player/corp wallets which *do not bear interest*. This is at least worrisome because it means that traditional monetary easing ("ISK printing") can not be trusted to have a stimulating fiscal effect (again similar to concerns surrounding the QE funds held by banks [LINK](http://www.cnbc.com/2013/11/11/the-problem-with-fed-qe-banks-just-arent-lending.html) [LINK](https://en.wikipedia.org/wiki/Liquidity_trap) [LINK](http://web.mit.edu/krugman/www/japtrap.html) etc. From CCP's perspective, a particularly troubling interpretation of this is that economic growth will become more expensive in terms of the new EVE subscribers required. 

At EVE Vegas 2015, taking advantage of modern computing power, CCP Quant presents a large-scale update of the EVE macroeconomy while also annoucing the launch of the new Monthly Economic Reports. [LINK](https://community.eveonline.com/news/dev-blogs/eve-economy-update-eve-vegas-2015-report/) Approximately half the report is about inflation, including a brief mention of PLEX prices. However, CCP Quant seems to go out of their way  to avoid any normative discussion about the CPI numbers which by themselves indicate no material change in the rate of inflation. The silence is poignant, especially after the defense of subscription numbers that opens the report. 

While this is only a brief excerpt of the economic research graciously compiled and published over the years, it does suggest that ISK deflation (and more specifically, how it has been addressed in the past) has become a key economic concern for CCP. 

It is worth briefly touching on the contemporaraneous RMT ("real-money-transaction") scandal that CCP has been addressing (there are other sources that can provide much more detail [LINK](http://nosygamer.blogspot.ca/)). Due to the fact that USD/ISK transactions are explicitly prohibited (traders must purchase primary-market PLEX or PLEX-denominated commodities to sell for ISK on CCP controlled markets), the only way to legally sell an ISK position for USD (or any other currency for that matter) is to redeem it for game time. The differnce in the USD subscription price and the USD/PLEX rate represents a frictional transaction cost that can be modeled against a risk premium for black market PLEX or ISK. 


# The Skill Point Economy

In February of 2016, CCP institutes its skill trading program by creating two new items, Skill Extractors ("extractors") and their output counterpart, Skill Injectors ("injectors"). This program completes the commodification of the time value of the skill points ("SP") that are paid out to any subscribed character with an active training queue; in other words, active character subscriptions are now interest-bearing. This has the effect of *lowering* the (nominally 0%) interest rate paid on wallet ISK -- in other words, this is a monetary easing program with the apparent goal of staving off deflation by increasing money velocity while holding the number of active subscribers constant. By adding or subtracting skill requirements, CCP can influence the demand side of the market. This was recently attempted
[LINK](https://forums.eveonline.com/default.aspx?g=posts&m=6326165) and abandoned after overwhelmingly negative feedback about the scale of the transition. 

## Assumptions and Definitions

To better illustrate the financial economics of CCP's new monetary policy, it is helpful to make (and therfore essential to state) some simplifying assumptions:

* Trader's entire balance at time $t = 0$ is held in ISK and and ISK is trader's preferred base currency.
* Trader's preferred secondary currency is USD
* Trader can effortlessly exchange USD for PLEX at a linear/fixed-per-unit rate from CCP (no economy of scale)
* Trader can effortlessly exchange USD for AUR at a linear/fixed-per-unit rate from CCP (no economy of scale)
* Trader can effortlessly exchange PLEX for AUR at a linear/fixed-per-unit rate from CCP (no economy of scale)
* Trader has cheap ISK credit in the very short run (i.e. for short intervals, skill extractors may be financed more or less for free)
* Trader has a "farming alt," a character with enough SP that they can use several extractors
* No price is 0


and importantly, for the time being,

* Trader abides by CCP Terms of Service (no secondary market)
***

While USD and ISK are both (capital controlled) currencies and SP is a non-monetary token, PLEX is actually a more complicated hybrid security/derivative that resembles a bond option [LINK](https://en.wikipedia.org/wiki/Bond_option) or more specifically, a perpetual [LINK](https://en.wikipedia.org/wiki/Perpetual_bond) floating rate note [LINK](https://en.wikipedia.org/wiki/Floating_rate_note) with an embedded swaption [LINK](https://en.wikipedia.org/wiki/Swaption). That is, the owner of a PLEX has the right, but not the obligation, to exchange the PLEX for a bond/swap that pays an ISK interest rate (historically, 0%). Of course the owner can hold the PLEX for the length of the term and sell it back for ISK -- in fact this is the rational choice when the expected ISK interest income is lower than the price increase of the PLEX. 

***
As a final note, while it is true that exercising a PLEX could in fact yield significant ISK income (either via PVE or PVP activities), it would almost always take additional capital to actually achieve a real return; so instead of considering this to be an interest rate on capital, we may instead think of this to be a yield on labor. The implications of this on modeling large scale aggregate demand are considerable but beyond the present scope of this work (and the author's expertise).


## Pricing

Consider a trader who holds an ISK balance of $B_t=S_{ISK/PLEX}$ at any starting time $t$ where $S_{ISK/PLEX}$ is the spot ISK price of 1 PLEX. The trader can use this balance to 
* hold as ISK (historically 0% yield)
* purchase risk assets (historically unknown yield)
* purchase a PLEX to hold and re-sell $T = 30 \text{ days}$ later (historically ~$2\%$ yield)
* purchase a PLEX to exercise for $T = 30 \text{ days}$ with the sole purpose of farming SP (new yield after skill trading introduced)


We can start with a simple case, one wherein investments only exist on the $T =  30 \text{ days}$ horizon (no compounding within months, no month-to-month investments).

***


### Exercising PLEX

Exercising PLEX produces a cashflow of $$\frac{SP}{hr} \cdot \frac{hr}{T} \cdot \frac{1}{\text{SP extraction cap}} \cdot E^{t+T}[p_{ISK}(\text{Injector})-p_{ISK}(\text{Extractor})] = k\cdot E^{t+T}[p_{ISK}(\text{Injector})-p_{ISK}(\text{Extractor})]$$

where $E^{t+T}[p_{ISK}(X)]$ is the forward price of X (in ISK) at time $t+T$ and $k$ is the number of skill injectors produced and sold over the course of $T$.

In this case, the one-period yield is $$y_T =  \frac{k \cdot (E^{t+T}[p_{ISK}(\text{Injector})]-E^{t+T}[p_{ISK}(\text{Extractor})])}{P_{ISK}(PLEX)}-1$$ where $P_{ISK}(PLEX)$ is the nominal price of the exercised PLEX.

We have not placed much significance on the expectation aspect of the notation of the forward price $E^{t+T}[p_{ISK}(X)]$; the above definitions are true as long as $E^{t+T}[p_{ISK}(X)] := p^{t+T}_{ISK}(X)$. Whether this value is predictable at all is a function of our probability measure and pricing framework. 

For the time being, let us assume that we can develop Martingale prices [LINK](https://en.wikipedia.org/wiki/Martingale_pricing) for every asset based on a risk-neutral probability measure (note that this entails a  no-arbitrage argument [LINK](http://www.columbia.edu/~mh2078/FE04/mtgale_pricing.pdf), which we will see is *not* the case under CCP capital controls). 



Then we can let $PV$ designate the present value of the cashflows following from exercising the PLEX:
$$PV=k\cdot [\frac{E^{t+T}[p_{ISK}(\text{Injector})]}{1+y} - \frac{E^{t+T}[p_{ISK}(\text{Extractor})]}{1+y}]$$


This only covers cases where we are restricted to purchasing Extractors with ISK. However, CCP's New Eden Store sells Extractors for AUR, which has a (historically) fixed conversion rate from PLEX.  Assuming that the AUR price for 1 PLEX stays constant over time ($\gamma := p_{AUR}(\text{PLEX})$ such that $\frac{\partial \gamma}{\partial t}\equiv 0$)

$$PV=k\cdot [\frac{E^{t+T}[p_{ISK}(\text{Injector})]}{1+y} - \frac{\min(\gamma \cdot E^{t+T}[p_{ISK}(\text{PLEX})],E^{t+T}[p_{ISK}(\text{Extractor}))]}{1+y}]$$

Then if $\gamma \cdot E^{t+T}[S_{ISK/PLEX}] \le  E^{t+T}[P_{ISK}(\text{Extractor})]$,

$$PV = k\cdot [\frac{E^{t+T}[p_{ISK}(\text{Injector})]}{1+y} - \frac{\gamma \cdot (1+r_{PLEX})\cdot p_{ISK}(\text{PLEX})}{1+y}]$$


*If the market is complete* ($\gamma \cdot E^{t+T}[S_{ISK/PLEX}] =  E^{t+T}[P_{ISK}(\text{Extractor})]$ via no-arbitrage): $$PV=k\cdot [\frac{E^{t+T}[p_{ISK}(\text{Injector})]}{1+y} - \frac{E^{t+T}[p_{ISK}(\text{Extractor})]}{1+y}]=k\cdot [\frac{E^{t+T}[p_{ISK}(\text{Injector})]}{1+y} - \frac{\gamma \cdot (1+r_{PLEX})\cdot p_{ISK}(\text{PLEX})}{1+y}] $$

then

$$r_{PLEX} = \frac{1}{\gamma}\cdot \frac{p_{ISK}(\text{Extractor})}{p_{ISK}(\text{PLEX})} - 1$$

where $r_{PLEX}$ is the PLEX return (from $t$ to $T$) that would be required ceteris paribus to make PLEX-priced Extractor financing priced fairly relative to ISK-priced Extractor financing.

It is worth noting that this is true even when $r_{PLEX} = y$, as expected.

The real-world significance of this fictitious Martingale price is that it quantifies the risk premium (and the corresponding liquidity impact of CCP's capital controls) that rewards or punishes PLEX investors. 

### Convexity: A Brief Note

If intra-periodic compounding is considered (i.e. injectors are produced, sold, and re-invested into other characters as soon as possible, not at the end of the subscription period), the present value of the cashflows can be treated more directly like a bond:


$$PV =  \Sigma_{j=1}^{k}{\frac{E^{j}[p_{ISK}(\text{Injector})]-\min(\gamma \cdot E^{j}[S_{ISK/PLEX}],E^{j}[p_{ISK}(\text{Extractor}))]}{(1+y)^j}}$$

but due to the extra assumptions required, we will not consider this in detail here.

# Empirical Results

```python
from IPython.display import *
import numpy as np
import seaborn as sns
%matplotlib inline 
from pandas import * 
import pycrest
type_id_dict = {'PLEX':29668,'Skill Injector':40520,'Skill Extractor':40519}
```

```python
eve = pycrest.EVE()
```

```python
type_id_dict = {'PLEX':29668,'Skill Injector':40520,'Skill Extractor':40519}

def crest_quote(name, eve = None, side = None, volume = 1, prices_only = True, region_id = 10000002):
    # Modified from https://github.com/flesser/django-crest-example
    global type_id_dict
    eve =  pycrest.EVE() if eve is None else eve
#     volume = int(volume)
    
    
    endpoint = eve._public_endpoint
    type_id = type_id_dict[name]        
    type_url = "{0}types/{1}/".format(endpoint, type_id)
    bids_url = "{0}market/{1}/orders/buy/?type={2}".format(endpoint, region_id, type_url)
    asks_url = "{0}market/{1}/orders/sell/?type={2}".format(endpoint, region_id, type_url)

    asks = eve.get(asks_url)['items']
    bids = eve.get(bids_url)['items']

    asks = sorted(asks, key=lambda k: k['price'])
    bids = sorted(bids, key=lambda k: k['price'], reverse=True)



    book = {'asks': asks, 'bids': bids}
    
    bidL2 = 1
    while sum([b['volume'] for b in book['bids'][:bidL2]]) < volume:
        bidL2 += 1
        if bidL2 >= len(book['bids']):
            bidL2 = 0
            break
    
    askL2 = 1
    while sum([a['volume'] for a in book['asks'][:askL2]]) < volume:
        askL2 += 1
        if askL2 >= len(book['asks']):
            askL2 = 0
            break
    bid_quote = book['bids'][:bidL2] if bidL2 > 0 else [{'price': np.nan}]
    ask_quote = book['asks'][:askL2] if askL2 > 0 else [{'price': np.nan}]
    
    return {'volume':volume,'bid': bid_quote[-1], 'ask':ask_quote[-1]} if not prices_only else {'volume':volume,'bid': bid_quote[-1]['price'],'ask': ask_quote[-1]['price']}


def history(name = None, type_id = None, days = 100, region_id = 10000002):
    global type_id_dict
    if name is None and type_id is None:
        return None
    elif type_id is None:
        type_id = type_id_dict[name]
    
    history = eve.get('https://public-crest.eveonline.com/market/{}/types/{}/history/'.format(region_id, type_id))
    out = history['items']
    while 'next' in history.keys():
        history = eve.get('https://public-crest.eveonline.com/market/{}/types/{}/history/'.format(region_id, type_id))
        out.extend(history['items'])
    out = DataFrame(out)
    return out.set_index(DatetimeIndex(out['date'])).drop('date',axis=1).iloc[-days:]
    
```

```python
sp_hr = 2700.
k = 24.*30.*sp_hr/500000.
N = 10
plex_aur = 3500.
extractor_aur = 1000.
gamma = extractor_aur / plex_aur
# plex_quote = crest_quote('PLEX',volume=N)
# inj_quote = crest_quote('Skill Injector',volume = N*k)
# ext_quote = crest_quote('Skill Extractor',volume = N*k)
```

First, we examine the recent price and volume series of PLEX, Injectors, and Extractors:

```python
region_id = 10000002
sns.plt.rcParams['figure.figsize'] = (15.0, 10.0);
hist_plex = history(name='PLEX', days = 100)
hist_inj = history(name='Skill Injector', days = 100)
hist_ext = history(name='Skill Extractor', days = 100)

sns.plt.plot(hist_plex['avgPrice'],'.');
sns.plt.plot(hist_inj['avgPrice'],'r.');
sns.plt.plot(hist_ext['avgPrice'],'k.');
sns.plt.legend(['PLEX','Skill Injector','Skill Extractor'],loc='best');
sns.plt.ylabel('Daily ISK Price');
sns.plt.title('Historical Average Daily ISK Prices');

sns.plt.figure();
sns.plt.plot(hist_plex['volume'],'.');
sns.plt.plot(hist_inj['volume'],'r.');
sns.plt.plot(hist_ext['volume'],'k.');
sns.plt.legend(['PLEX','Skill Injector','Skill Extractor'],loc='best');
sns.plt.ylabel('Daily Volume');
sns.plt.title('Historical Daily Trading Volume');

```

After the initial frenzy of extraction on and around the inception date, trading volume in PLEX has decayed predictably to a higher baseline than did exist before the inception.

Of particular note here is the decline in PLEX prices. The parabolic drops in late March reflect a significant amount of selling pressure; the bullish activity following the first of these seems to either reflect a tepid reversion trade *or* market intervention by CCP's trading desk. Regardless, PLEX prices continue to decline.

Using the model outlined above, we can price our risk-neutral $r_{PLEX}$:

```python
sns.plt.plot(1/gamma*np.divide(hist_ext['avgPrice'],hist_plex['avgPrice'].loc[hist_ext.index])-1,'o-');
sns.plt.title('Implied Risk-Neutral PLEX Return (Based on Daily Average Prices)');
sns.plt.ylabel('Return');
```

The negative return could be interpeted to say that **people are getting rid of their PLEX faster than they are getting rid of their ISK**.

While this value has been disctintly negative since the inception of skill trading, the fact that the price has sunk in a largely unbroken trend is even more troubling. Whether it was investor activity or CCP intervention that yielded the modest recovery in late March, those gains have been wiped out **over the course of 2 days** at the time of writing.

The extraction spread shown here is the payoff of extracting and marketing a single skill injector in real-time:

```python
sns.plt.plot((hist_inj['avgPrice']-gamma*hist_plex['avgPrice'].loc[hist_ext.index]),'o-');
sns.plt.ylabel('ISK');
sns.plt.plot((hist_inj['avgPrice']-hist_ext['avgPrice']),'o-');
sns.plt.legend(['PLEX-based','ISK-based']);
sns.plt.title('Extraction Spread (Based on Daily Average Spot Prices)');
```

Clearly, it has always been more profitable in nominal terms to finance extraction in ISK. 




Since skill trading inception, the PLEX price of extractors has consistently traded at a premium over the ISK price. 

```python
sns.plt.plot(gamma*hist_plex['avgPrice'].loc[hist_ext.index] - hist_ext['avgPrice'],'o-');
sns.plt.legend(['PLEX Price - ISK Price']);
sns.plt.ylabel('ISK');
sns.plt.title('PLEX Premium on Extractors (Daily Average Prices)');
```

This is a peculiar state of affairs because Skill Extractors are not producible by any market participant except CCP. Either the extractors on the ISK market are being sold at a (real) loss by purchasers or CCP is flooding the ISK market with (nominally discounted) extractors in a misguided attempt to disincentivize the flow of PLEX into Skill Extractors. The implication is that a PLEX purchaser/investor can achieve a higher return by exercising their PLEX and financing extractor purchases exclusively in ISK. Indeed, this would incentivize PLEX purchases across the board and, as quoted above,

>PLEX seems to increase the velocity of money in the New Eden economy as it "liberates" ISK from people with more money than they know what to do with and gets it to ISK needy people that create and sell PLEX

However, it would appear that whatever demand has been stimulated by this mechanism has been absolutely overwhelmed by the volume of PLEX being liquidated for ISK.

The missing piece is that PLEX can be licitly purchased from CCP for USD. Although PLEX sales are not publicised by CCP, it would appear that the volume of PLEX purchased for USD (and subsequently sold for ISK) is **far** outpacing the volume of PLEX purchased and exercised (either expressly for skill farming or for any other purpose) to the point that the market is putting a *significant* risk premium on ISK. 

One might speculate that this ISK is primarily being used to purchase Skill Injectors from producers and so is in fact stimulating the broader economy; regardless, this capital does not seem to be re-invested into the PLEX market at a rate that stabilizes PLEX prices. If that is the case, than the seniorage profit that CCP earns by selling PLEX on the primary market represents **the flow of potentially productive capital out of the ISK economy and onto CCP's balance sheet.** Time will tell.

# Conclusions and Future Work

With the clarity of hindsight, we can all too easily call into question the CCP policies that have directed the ISK economy for the last 6 years. Indeed, Eyjólfur's assumption that the exploding money supply was likely to lead to a hyper-inflationary spiral appears to have been made (or at least stated) without any consideration of the velocity of ISK. The recognition of the deflationary pattern of the ISK economy was an important realization and the skill trading/time-value concept is a competent response in the absence of any other way to boost subscription numbers. However, the execution of this plan has only placed more significance on the market frictions caused by CCP's strict capital controls and ban on RMT/secondary trading. 

Although it is too early to judge the long-term consequences of the program, unless CCP can stabilize PLEX prices (e.g. by raising the USD price of PLEX or by bidding up the ISK price of extractors/injectors), **there is a serious danger of exacerbating and accelerating the problem this program was designed to solve.** One would hope that after years of surveying the data and literature, CCP has realized that the problem is not the glut of the ISK money supply or even the low velocity of money; these are only symptoms of the sustained lack of real aggregate demand.
